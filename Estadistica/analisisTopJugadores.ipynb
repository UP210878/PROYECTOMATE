```python
path = '../batting.csv'
df_bat = pd.read_csv(path, sep = ',')
```

```python
df_finalbat = df_bat.merge(df_clean2,'inner','player_id')
```

```python
df_finalbat.head(20)
```

```python
df_finalclean = df_finalbat.copy()
```

```python
df_finalclean.drop_duplicates(subset='player_id',keep='last',inplace=True)
```
